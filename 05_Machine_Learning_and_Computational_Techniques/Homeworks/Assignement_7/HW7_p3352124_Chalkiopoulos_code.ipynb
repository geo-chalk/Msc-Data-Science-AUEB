```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math


# Plot the data in pairs to see the distribution
fig = plt.figure(figsize = (10, 7))
ax = plt.axes()


mu = -1
variance = 1
sigma = math.sqrt(variance)
x1 = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
ax.plot(x1, stats.norm.pdf(x1, mu, sigma), label='$\omega_1$')

mu_2 = 1
variance_2 = 4
sigma_2 = math.sqrt(variance_2)
x2 = np.linspace(mu_2 - 3*sigma_2, mu_2 + 3*sigma_2, 100)
ax.plot(x2, stats.norm.pdf(x2, mu_2, sigma_2), label='$\omega_2$')
ax.set_ylim(0,0.42)
ax.set_xlim(-6,8)

ax.axvspan(-3.57, 0.24, facecolor='cyan', alpha = 0.4, label='$R_1$')
ax.axvspan(-6, -3.57, facecolor='peru', alpha = 0.4, label='$R_2$')
ax.axvspan(0.24, 8, facecolor='peru', alpha = 0.4)


ax.legend()
```

```python
mean_1 = np.array([[6],[0]])
mean_2 = np.array([[0],[6]])
cov = np.array([[2,0],[0,2]])
X = np.array([[2,4,4,2,2,2]]).reshape(3,2)
p1 = 0.5
p2 = 0.5
```

```python

for i, x in enumerate(X):


    cov_norm_1, cov_norm_2 = np.linalg.norm(cov)**1/2, np.linalg.norm(cov)**1/2
    cov_inv_1, cov_inv_2 = np.linalg.inv(cov), np.linalg.inv(cov)
    x_mu_1, x_mu_2 = x.reshape(-1,1) - mean_1, x.reshape(-1,1) - mean_2

    p_x_1 = 1/ ( (2*np.pi) * cov_norm_1 ) * np.exp(-0.5 * (x_mu_1.T).dot(cov_inv_1).dot((x_mu_1)))
    p_x_2 = 1/ ( (2*np.pi) * cov_norm_2 ) * np.exp(-0.5 * (x_mu_2.T).dot(cov_inv_2).dot((x_mu_2)))

    P1_x = p1 * p_x_1 / (p1 * p_x_1 + p2 *p_x_2)
    P2_x = p2 * p_x_2 / (p1 * p_x_1 + p2 *p_x_2)

    print(f"p(1 | x_{i+1}): ", f"{p_x_1[0][0]:.2e}")
    print(f"p(2 | x_{i+1}): ", f"{p_x_2[0][0]:.2e}")
```

```python
x = np.linspace(-4,4,1000)
p1 = 1 / (np.sqrt(2*np.pi)) * np.exp(-x**2/2)

p2 = []
for i in x:
    if abs(i) < np.sqrt(2*np.pi):
        p2.append(1 / (2*np.sqrt(2*np.pi)))
    else:
        p2.append(0)
```

```python
# Plot the data in pairs to see the distribution
fig = plt.figure(figsize = (10, 7))
ax = plt.axes()

ax.plot(x, p1, label='$p(x \\vert \omega_1)$')
ax.plot(x, p2, label='$p(x \\vert\omega_2)$')
ax.legend()
_ = ax.set_ylim(0,0.42)
```
